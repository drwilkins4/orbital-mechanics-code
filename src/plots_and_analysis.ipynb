```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt
import f90nml

plt.style.use('fivethirtyeight')
```

```python
# reading data stored in 2planets_case1_output.dat

file_name = '2planets_case1_output.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# Use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# Transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

t = all_data[0]
x1 = all_data[1]
y1 = all_data[2]
x2 = all_data[3]
y2 = all_data[4]
E = all_data[5]
```

```python
# Plotting case 1. Two distant planets in circular orbit around primary mass.

plt.plot(x1, y1, label="Planet 1")
plt.plot(x2, y2, label="Planet 2")

# Set appropriate labels with the correct units.
plt.xlabel('X')
plt.ylabel('Y')
# legend displays the labels given in each plot
plt.legend()
plt.show()
plt.show()

# print initial conditions
parameters1 = f90nml.read('2planets_case1.namelist')
x_0_1 = parameters1['initial_conditions']['initial_pos_1']
v_0_1 = parameters1['initial_conditions']['initial_vel_1']
x_0_2 = parameters1['initial_conditions']['initial_pos_2']
v_0_2 = parameters1['initial_conditions']['initial_vel_2']
t_f = parameters1['solution_parameters']['final_time']
n = parameters1['solution_parameters']['n_steps']
print("Planet 1 initial position: ", x_0_1)
print("Planet 1 initial velocity: ", v_0_1)
print("Planet 2 initial position: ", x_0_2)
print("Planet 2 initial velocity: ", v_0_2)
print("Final time: ", t_f)
print("Number of steps: ", n)
```

```python
# Plotting energy vs. time for case 1

plt.plot(t, E, label='Energy vs Time')

# Set appropriate labels with the correct units.
plt.xlabel('Time')
plt.ylabel('Energy')
# legend displays the labels given in each plot
plt.legend()
plt.show()
plt.show()
```

At first glance it appears as though the energy decreases drastically. However, the change from -10000 to -10100 is negligible. It is the scaling of the graph that makes it appear this way.
